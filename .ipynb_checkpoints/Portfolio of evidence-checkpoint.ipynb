```python
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize

# Opening the file for future purposes
df = open(r'C:\Users\shane\Documents\GitHub\Data-Analytics-Repository\Portofolio of Evidence\un_declaration_hr_text_data.txt')

# Reading the file's contents
text = df.read()
blob = TextBlob(text)

#print(blob)

from collections import Counter

stops = stopwords.words('english')
custom_stopwords = [
    'is', 'in', 'not', 'to', 'and', 'it\'s', 'while', 'they', 'at', 'what', 'like', 'Your', 'are', 'here', 'over', 'as', 'opens',
    'the', 'A', 'a', 'I', 'it', 'on', 
]
stops = stops + custom_stopwords

# Tokenizing the text
tokens = word_tokenize(text)

# Removing the stopwords
words = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stops]

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(words))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top 100 Words Word Cloud')
plt.tight_layout()
plt.show()

# Counting the occurrences
term_frequencies = Counter(words)

# Getting the top 25 frequent items
top_terms = term_frequencies.most_common(25)

# Plotting the bar graph
terms, frequencies = zip(*top_terms) # Unpacking the tuple
plt.figure(figsize=(12,9))
plt.barh(range(len(terms)), frequencies, align='center')
plt.yticks(range(len(terms)), terms)
plt.xlabel('Terms')
plt.ylabel('Frequency')
plt.title('Frequency of Terms in the UN Declaration')
plt.show()
```
